## 5. Import the pandas library we just installed into the Python environment, along with boto3 and io libraries


```python
import pandas as pd
import boto3
import io
```

##  6. Read the .csv file from the S3 bucket into a dataframe using pandas

```python

bucket = 'hfg2022-demo-folder'
s3_file_key = 'titanic.csv' # this is the file name, but you may also need to add any folder names in here

# Create a boto3 client to allow SageMaker to talk to S3
s3 = boto3.client('s3')

# Get a handle to the S3 .csv file
fileobj = s3.get_object(Bucket=bucket, Key=s3_file_key)

# Read the .csv file into a pandas data frame
dataframe = pd.read_csv(io.BytesIO(fileobj['Body'].read()))
```
